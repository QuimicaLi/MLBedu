```python
fecha_2 = pd.to_datetime(delitos_data['fecha_inicio'], format="%Y-%m-%d")
fecha_2.min(), fecha_2.max()
```

```python
#¿Cómo se distribuye el número de delitos en la CDMX?
#¿Cuáles son los 5 delitos más  frecuentes?
# Delitos más frecuentes
delitos_data['fecha'] = fecha_2
delitos_frecuentes = delitos_data['delito'].value_counts().nlargest(8).index.tolist()
delitos_data_filtered = delitos_data[(delitos_data['fecha'].dt.year <= 2023) & (delitos_data['delito'].isin(delitos_frecuentes))]

pivot_table = delitos_data_filtered.groupby(delitos_data_filtered['fecha'].dt.strftime('%b'))['delito'].value_counts().unstack()

# Ordenar los meses correctamente
meses_ordenados = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pivot_table = pivot_table.reindex(meses_ordenados, axis=0)

pivot_table.plot(kind='bar', stacked=True)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Generar columna de mes y año
delitos_data['fecha'] = pd.to_datetime(delitos_data['fecha_inicio'])
delitos_data['mes'] = delitos_data['fecha'].dt.strftime('%b')
delitos_data['anio'] = delitos_data['fecha'].dt.year

# Filtrar los delitos más frecuentes hasta 2023
delitos_frecuentes = delitos_data['delito'].value_counts().nlargest(8).index.tolist()
delitos_data_filtered = delitos_data[(delitos_data['anio'] <= 2023) & (delitos_data['delito'].isin(delitos_frecuentes))]

# Agrupar por alcaldía y delito
pivot_table = delitos_data_filtered.groupby(['alcaldia_hecho', 'delito']).size().unstack()

# Graficar
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Número de delitos por alcaldía')
plt.xlabel('Alcaldía')
plt.ylabel('Número de delitos')
plt.legend(title='Delito', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Agrupar por alcaldía y contar el número de delitos en cada una
delitos_por_alcaldia = delitos_data.groupby('alcaldia_hecho').size().sort_values(ascending=False)

print("Distribución del número de delitos por alcaldía:")
print(delitos_por_alcaldia)
# Graficar la distribución del número de delitos por alcaldía
plt.figure(figsize=(10, 6))
delitos_por_alcaldia.plot(kind='bar', color='skyblue')
plt.title('Distribución del número de delitos por alcaldía')
plt.xlabel('Alcaldía')
plt.ylabel('Número de Delitos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Analizamos algunos de los delitos más frecuentes, para identificar tendencia de alza o baja con una serie de tiempo.
fecha_3 = pd.to_datetime(delitos_data['fecha_hecho'], format="%Y-%m-%d")
delitos_data['fecha_2023'] = fecha_3
delitos_anio_anterior = delitos_data[delitos_data['fecha_2023'].dt.year == 2023]
delitos_anio_anterior_filtered = delitos_anio_anterior[delitos_anio_anterior['delito'].isin(delitos_frecuentes)]
conteo_mes = delitos_anio_anterior_filtered.groupby([delitos_anio_anterior_filtered['fecha_2023'].dt.month.rename('mes'), 'delito']).size().reset_index(name='n')

pivot_table = conteo_mes.pivot(index='mes', columns='delito', values='n')
pivot_table.plot()

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
